```python
import plotly
import pandas as pd
import plotly.graph_objects as go
```

```python
data = pd.read_csv("data/nz_weather.csv")
data.head()
```

```python
fg1 = go.Scatter(x=data['DATE'],
                 y=data['Auckland'],
                 name='Auckland',
                line={'color':'red'}
                
                )
fg2 = go.Scatter(x=data['DATE'],
                 y=data['Wellington'],
                 name='Wellington',
                line={'color':'blue'}
                )
fg = go.Figure((fg1,fg2))
fg.update_layout(
    title='my graph',
    xaxis_title='date',
    yaxis_title='weather',
    template='plotly_dark',
    xaxis={'color':"blue"},
    yaxis={'color':"blue"}
)
fg.show()
```

```python
data10 = data[(data['DATE']>='2010-01') & (data['DATE']<='2010-12')]
data10
```

```python
bar1 = go.Bar(x=data10['DATE'],y=data10['Auckland'],name='Auckland',text=data10['Auckland'],textposition='outside')#最后的参数是是否将数字显示在柱子上面
fg = go.Figure(bar1)
fg.show()
```

```python
hist = go.Histogram(x=data['Auckland'],xbins={'size':20})
fig = go.Figure(hist)
fig.update_layout(bargap=0.1)
fig.show()
```
